```python
import numpy as np
import matplotlib.pyplot as plt
```

# Desplazamientos verticales y horizontales

```python
# 5 unidades hacia arriba -> f(x) + c
x = np.linspace(-10,10,1000)

def f(x):
    return x**2

y = f(x) + 5

plt.plot(x,y)
plt.axhline(y=0,color='r')
plt.axvline(x=0,color='r')
plt.grid()
plt.show()
```

```python
#5 unidades hacia abajo -> f(x) - c
y = f(x) - 5

plt.plot(x,y)
plt.axhline(y=0,color='r')
plt.axvline(x=0,color='r')
plt.grid()
plt.show()
```

```python
#5 unidades hacia la derecha -> f(x - c) 
y = f(x - 5) 

plt.plot(x,y)
plt.axhline(y=0,color='r')
plt.axvline(x=0,color='r')
plt.grid()
plt.show()
```

```python
#5 unidades hacia la izquierda -> f(x + c) 
y = f(x + 5) 

plt.plot(x,y)
plt.axhline(y=0,color='r')
plt.axvline(x=0,color='r')
plt.grid()
plt.show()
```

# Alargamientos y compresiones

```python
#Alargamiento vertical -> f(x) * c
x = np.linspace(-15,15,1000)

def f(x):
    return np.sin(x)

y = f(x) * 2

plt.plot(x,y)
plt.axhline(y=0,color='r')
plt.axvline(x=0,color='r')
plt.grid()
plt.show()
```

```python
#Comprension vertical -> f(x) * 1/c 
y = f(x) * 1/2

plt.plot(x,y)
plt.axhline(y=0,color='r')
plt.axvline(x=0,color='r')
plt.grid()
plt.show()
```

```python
#Alargamiento horizontal -> f(x * 1/c ) 
y = f(x * 1/2)

plt.plot(x,y)
plt.axhline(y=0,color='r')
plt.axvline(x=0,color='r')
plt.grid()
plt.show()
```

```python
#Comprension horizontal -> f(x * c ) 
y = f(x * 2)

plt.plot(x,y)
plt.axhline(y=0,color='r')
plt.axvline(x=0,color='r')
plt.grid()
plt.show()
```

# Reflexiones

```python
#Reflexion eje x -> -f(x)
def f(x):
    return x**3

y = -f(x)

plt.plot(x,y)
plt.axhline(y=0,color='r')
plt.axvline(x=0,color='r')
plt.grid()
plt.show()
```

```python
#Reflexion eje y -> f(-x)
y = f(-x)

plt.plot(x,y)
plt.axhline(y=0,color='r')
plt.axvline(x=0,color='r')
plt.grid()
plt.show()
```
