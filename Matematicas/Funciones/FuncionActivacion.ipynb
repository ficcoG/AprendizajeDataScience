```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10,10,1000)
```

# Función Lineal

```python
#Para mantener valores a lo largo de un proceso  -> Tendencias
def f(x):
    return x
y = f(x)

plt.plot(x,y)
plt.show()
```

# Función escalon o de Heaviside

```python
#Valores < 0 -> 0
#Valores > 0 -> 1

#Clasificaciones categoricas -> Modelos binarios

def h(x):
    Y = np.zeros((len(x)))
    for idx,x in enumerate(x):
        if x>=0:
            Y[idx]=1
    return Y

y = h(x)

plt.plot(x,y)
plt.show()
```

# Función Sigmoide

```python
#Regresa valores entre 0 y 1
#Valor 0 -> Regresa 0.5
#Regresion logistica -> Probabilidades
#Valores grandes tienden a 0 o 1 dependiendo el signo

def f(x):
    return 1/(1 + np.exp(-x))

y = f(x)

plt.plot(x,y)
plt.grid()
plt.show()
```

# Función Tangente Hiperbólica

```python
#Regresa valores entre -1 y 1
#Escalamiento
#Valor 0 -> Regresa 0

def f(x):
    return np.tanh(x)

y = f(x)

plt.plot(x,y)
plt.grid()
plt.show()
```

# Función ReLu

```python
#Valores < 0 -> Regresa 0
#Valores > 0 -> Regresa x
#Simulacion de "neuronas muertas" 
def f(x):
    return np.maximum(x,0)

y = f(x)

plt.plot(x,y)
plt.grid()
plt.show()
```
