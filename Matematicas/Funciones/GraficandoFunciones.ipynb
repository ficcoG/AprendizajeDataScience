```python
import numpy as np
import matplotlib.pyplot as plt
```

# Funciones Algebraicas

## Funci√≥n Lineal

```python
N = 100
m = 10
b = 5

def f(x):
    return m*x+b

x = np.linspace(-10,10,num=N)

y=f(x)

fig, axe = plt.subplots()
axe.plot(x,y)
plt.show()
```

## Funciones Polin√≥micas

```python
#Grado impar
def f(x):
    return (2*x**7)-(x**4)+(3*x**2)+4

y=f(x)

plt.plot(x,y)
plt.show()
```

```python
#Grado par
def f(x):
    return (2*x**6)-(x**4)+(3*x**2)+4

y=f(x)

plt.plot(x,y)
plt.show()
```

# Funciones trascendentes

## Funciones trigonom√©tricas

```python
def f(x):
    return np.cos(x)

y = f(x)

plt.plot(x,y)
plt.show()
```

```python
def f(x):
    return np.sin(x)

y = f(x)

plt.plot(x,y)
plt.show()
```

```python
def f(x):
    return np.tan(x)

y = f(x)

plt.plot(x,y)
plt.show()
```

## Funci√≥n exponencial

```python
def f(x):
    return np.exp(x)

y = f(x)

plt.plot(x,y)
plt.show()
```

## Funci√≥n logar√≠tmica

```python
def f(h):
    return np.log2(h)

#Solo acepta valores mayores a 0

h = np.linspace(0.001,256,num=1000)

y = f(h)

plt.plot(h,y)
plt.show()
```

# Funci√≥n Seccionada

```python
def f(x):
    Y = np.zeros(len(x))
    for idx,x in enumerate(x):
        if x>=0:
            Y[idx]=1
    return Y
    

y = f(x)

plt.plot(x,y)
plt.show()
    
```

```python
#Funcion valor absoluto
def f(x):
    return np.absolute(x)

y = f(x)

plt.plot(x,y)
plt.show()
```
