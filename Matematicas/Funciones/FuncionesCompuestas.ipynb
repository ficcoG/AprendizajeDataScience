```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10,10,100)

def g(x):
    return x**2

def f(x):
    return np.sin(x)

y = g(x)

f_o_g = f(y)

plt.plot(x,f_o_g)
plt.show()
```

```python
#Otra manera de definirlo
f_o_g = f(g(x))

plt.plot(x,f_o_g)
plt.show()
```
