```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.histplot(data=tips,x='tip')
plt.show()
```

```python
#Acumulamos los valores
sns.histplot(data=tips,x='tip',cumulative=True)
plt.show()
```

```python
sns.histplot(data=tips,x='tip',hue='sex')
plt.show()
```

```python
sns.histplot(data=tips,x='tip',stat='frequency')
plt.show()
```

```python
sns.histplot(data=tips,x='tip',stat='probability')
plt.show()
```

```python
sns.histplot(data=tips,x='tip',stat='density')
plt.show()
```

```python
#Stackean las barras
sns.histplot(data=tips,x='tip',hue='sex',multiple='stack')
plt.show()
```

```python
#Uno al lado del otro
sns.histplot(data=tips,x='tip',hue='sex',multiple='dodge')
plt.show()
```

```python
#Fill, de 0 a 1 -> Mostrando la probabilidad
sns.histplot(data=tips,x='tip',hue='sex',multiple='fill')
plt.show()
```

# Kde -> Diagrama de densidad

```python
sns.kdeplot(data=tips,x='tip',hue='sex')
plt.show()
```

```python
sns.kdeplot(data=tips,x='tip',hue='sex',cumulative=True)
plt.show()
```

```python
#Pinta el area bajo la curva
sns.kdeplot(data=tips,x='tip',hue='sex',shade=True)
plt.show()
```

```python
sns.kdeplot(data=tips,x='tip',hue='sex',bw_adjust=1)
plt.show()
```

# Ecdf

```python
#Maneja la proporcion
sns.ecdfplot(data=tips,x='tip',hue='sex')
plt.show()
```

```python
sns.ecdfplot(data=tips,x='tip',hue='sex',stat='count')
plt.show()
```

# Diagrama de distribucion

```python
sns.displot(data=tips,x='tip',hue='sex')
plt.show()
```

```python
#Cambiamo el diagrama desde el parametro
sns.displot(data=tips,x='tip',hue='sex',kind='kde')
plt.show()
```
