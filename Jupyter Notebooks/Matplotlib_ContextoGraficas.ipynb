```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

# Contexto a las graficas

```python
fig, axes = plt.subplots(1,2,figsize=(5,5)) #Modificamos el tama√±o de la figura(Lienzo)
axes[0].plot(x,y,label='$sin(x)$') #Definimos la etiqueta del grafico ----Entre '$ $' lo definimos como funcion matematica
axes[0].set_title('Relacion X - Y') #Agregamos un titulo al grafico
axes[0].set_xlabel('X') #Agregamos una etiqueta al eje X
axes[0].set_ylabel('Y') #Agregamos una etiqueta al eje Y
axes[0].legend() #Mostramos la etiqueta del grafico

axes[1].plot(y,x)
axes[1].set_title('Relacion Y - X')
axes[1].set_xlabel('Y') 
axes[1].set_ylabel('X') 
plt.show()
```

```python
plt.plot(x,y,label='sin(x)')
plt.title('Este es un titulo')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='lower left') #Podemos configurar donde se ubica el label
plt.show()
```
