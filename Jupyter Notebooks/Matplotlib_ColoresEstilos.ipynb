```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
#Estilos disponibles
print(plt.style.available) 
```

```python
plt.style.use('seaborn-pastel')
```

## Matlab

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.plot(x,x+1,'r--')
ax.plot(x,x+2,'b-')
ax.plot(x,x+3,'y.-')
ax.plot(x,x+4,'g:')
```

## Pyplot

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.plot(x,x+1,color = 'green',alpha=0.5) #Alpha -> Transparencia
ax.plot(x,x+2,color = 'blue',linewidth=8) #Linewidth -> Grosor de las lineas
ax.plot(x,x+3,color = '#66FF66',linestyle = ':') #Linestyle -> Estilo de la linea
ax.plot(x,x+4,color = '#CCFFCC',marker = 'o',markersize=10,markerfacecolor = 'red') #
```
