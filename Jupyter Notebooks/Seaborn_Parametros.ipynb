```python
import seaborn as sns
import matplotlib.pyplot as plt
```

# Datasets cargados en seaborn

```python
#Para mostrar los datasets disponibles
sns.get_dataset_names()
```

```python
#Cargamos el dataset
tips = sns.load_dataset('tips') 
tips
```

# Parametros importantes -> (data,x,y,hue,kind)

```python
sns.displot(data=tips,x='total_bill')
plt.show()
```

```python
sns.displot(data=tips,x='total_bill',y='tip')
plt.show()
```

```python
#Hue para segmentar los datos
sns.displot(data=tips,x='total_bill',y='tip',hue='sex')
plt.show()
```

```python
sns.displot(data=tips,x='total_bill',hue='sex')
plt.show()
```

```python
sns.displot(data=tips,x='total_bill',hue='sex',kind='kde')
plt.show()
```

```python
sns.displot(data=tips,x='total_bill',hue='sex',kind='kde',legend=False)
plt.show()
```

```python
sns.displot(data=tips,x='total_bill',hue='sex',kind='kde',palette='dark',alpha=0.1
           )
plt.show()
```
