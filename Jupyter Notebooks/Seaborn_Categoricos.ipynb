```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

# Categoricos

# Countplot

```python
sns.countplot(data=tips,x='day',hue='sex')
plt.show()
```

```python
#De manera horizontal modificando el eje
sns.countplot(data=tips,y='day',hue='sex')
plt.show()
```

# Stripplot

```python
sns.stripplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True)
plt.show()
```

# Swarmplot

```python
#Modifico el tamao de los markers para mayor distincion
sns.swarmplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True,size=4.5)
plt.show()
```

# Boxplot

```python
sns.boxplot(data=tips,x='day',y='total_bill',hue='sex')
plt.show()
```

# Unimos los graficos

```python
sns.swarmplot(data=tips,x='day',y='total_bill',hue='sex',dodge=True,size=4.5)
sns.boxplot(data=tips,x='day',y='total_bill',hue='sex')
plt.show()
```

# Violinplot

```python
sns.violinplot(data=tips,x='day',y='total_bill',hue='sex')
plt.show()
```

```python
#Combinamos los graficos con split
sns.violinplot(data=tips,x='day',y='total_bill',hue='sex',split=True)
plt.show()
```

# Catplot -> permite utilizar todos los graficos categoricos

```python
sns.catplot(data=tips,x='day',y='total_bill',hue='sex')
plt.show()
```

```python
sns.catplot(data=tips,x='day',y='total_bill',hue='sex',kind='swarm')
plt.show()
```

```python
sns.catplot(data=tips,x='day',y='total_bill',hue='sex',kind='box')
plt.show()
```

```python
#Agrego una tercer variable con el parametro 'col'
sns.catplot(data=tips,x='day',y='total_bill',hue='sex',kind='swarm',col='time')
plt.show()
```
