```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

# Subplots

```python
fig, axes = plt.subplots()
axes.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y) #[0]---> Hace referencia a la primera posicion de la grafica
axes[1].plot(y,x,'r')
```

```python
fig, (ax1,ax2) = plt.subplots(nrows=1,ncols=2) #Defino previamente mediante una tupla a cada axes
ax1.plot(x,y) 
ax2.plot(y,x,'r')
```

```python
fig, axes = plt.subplots(2,4) # == que poner nrows y ncolumns
axes[0,0].plot(x,np.cos(x))
axes[0,1].plot(x,np.sin(x),'r')
axes[0,2].plot(x,np.tan(x),'y')
axes[0,3].plot(x,np.cos(x)**2)
```

```python
#Hacemos lo mismo pero usando tuplas
fig, ((ax1,ax2,ax3,ax4),(ax5,ax6,ax7,ax8)) = plt.subplots(2,4) 
ax1.plot(x,np.cos(x))
ax2.plot(x,np.sin(x),'r')
ax3.plot(x,np.tan(x),'y')
ax4.plot(x,np.cos(x)**2)
fig.tight_layout() #Para  separar las graficas entre si
```

```python
#Terminamos de completar el lienzo
fig, ((ax1,ax2,ax3,ax4),(ax5,ax6,ax7,ax8)) = plt.subplots(2,4) 
ax1.plot(x,np.cos(x))
ax2.plot(x,np.sin(x),'r')
ax3.plot(x,np.tan(x),'y')
ax4.plot(x,np.cos(x)**2)
ax5.plot(y,x,'g')
ax6.plot(y,np.cos(y),'y')
ax7.plot(y,np.tan(y))
ax8.plot(y,np.sin(y),'r')
fig.tight_layout() 
plt.show()
```
