```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.randint(1,50,100)
```

# Histograma

```python
plt.hist(data,bins=20,histtype = 'bar') #Numero de barras
plt.show()
```

```python
plt.hist(data,bins=20,histtype = 'step') #Tipo de histograma
plt.show()
```

# Boxplot

```python
#Grafico util para entender la distribucio de los datos
data = np.append(data,100) #El grafico muestra el outlayers, punto fuera de la concentracion de los datos
plt.boxplot(data,vert=False,patch_artist=True,notch=True) #(Girar el grafico,Rellenar el grafico,enfoca hacia la mediana
plt.show()
```

# Grafico de dispersion - Scatter

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
area = (30 * np.random.rand(N))**2
colors = np.random.rand(N)
```

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.scatter(x,y,s=area,c=colors,marker='o',alpha=0.5)
plt.show()
```
