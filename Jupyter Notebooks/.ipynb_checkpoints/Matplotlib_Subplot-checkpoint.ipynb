```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

# Subplot

```python
#Permite crear mas graficas
plt.subplot(1,2,1) # (Numero de filas,Numero de columnas,index)
plt.plot(x,y)
plt.subplot(1,2,2)
plt.hist(y)
plt.show()
#Obtenemos 1 fila y dos columnas
```

```python
plt.subplot(1,2,1) # (Numero de filas,Numero de columnas,index)
plt.plot(x,y,'b:')
plt.plot(y,x,'ro-') #Utiliza el mismo grafico que sigo parado en el mismo indice
plt.subplot(1,2,2)
plt.pie(y)
plt.show()
```

```python
plt.subplot(2,1,1) # (Numero de filas,Numero de columnas,index)
plt.plot(x,y)
plt.subplot(2,1,2)
plt.hist(y)
plt.show()
#Obtenemos 2 filas y 1 columna
```
