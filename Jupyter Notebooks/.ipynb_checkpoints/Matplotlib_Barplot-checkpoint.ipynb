```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
country = ['India','Japan','Mexico','Colombia','Argentina']
population = [1000,800,900,1000,300]
```

# Grafico de barras

```python
plt.bar(country,population,width=0.5,color=['olivedrab','yellowgreen','lightgreen','y','forestgreen'])
plt.title('Poblacion por pais')
plt.xlabel('Pais')
plt.ylabel('Poblacion')
plt.xticks(np.arange(5),('India','Japon','Mexico','Colombia','Argentina'),rotation=45)
plt.show()
```

```python
#Grafico de barras horizontal
plt.barh(country,population)
plt.show()
```
