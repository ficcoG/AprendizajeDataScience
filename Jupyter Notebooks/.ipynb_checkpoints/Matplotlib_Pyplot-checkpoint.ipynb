```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y= x**2
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'ro') # 'r' de rojo --- 'o' del marker
plt.show()
```

```python
#Otro ejemplo
plt.plot(x,y,'bx') # 'b' de azul --- 'x' del marker
plt.show()
```

```python
plt.plot(x,y,'bs-')# 'b' de azul --- 's' del marker(square)---- '-' para marcar una continuidad
plt.show()
```

```python
plt.hist(x)
plt.show()
```

```python
plt.pie(x)
plt.show()
```

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.boxplot(x)
plt.show()
```
