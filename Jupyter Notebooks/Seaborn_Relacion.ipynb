```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

# Scatterplot

```python
sns.scatterplot(data=tips,x='total_bill',y='tip')
plt.show()
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='day',alpha=0.5)
plt.show()
```

```python
#Modificamos el tamao de la figura
plt.figure(figsize=(12,7))
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='day',alpha=0.5,style='time',size='size')
#Movemos la leyenda
plt.legend(loc='center',bbox_to_anchor=(1.10,0.5))
plt.show()
```

```python
plt.figure(figsize=(12,7))
markers = {'Lunch':'D','Dinner':'s'}
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='day',alpha=0.5,style='time',size='size',markers=markers)
plt.legend(loc='center',bbox_to_anchor=(1.10,0.5))
plt.show()
```

# Lineplot

```python
sns.lineplot(data=tips,x='total_bill',y='tip')
plt.show()
```

# Pairplot

```python
sns.pairplot(data=tips,hue='day')
plt.show()
```

# Relplot -> permite utilizar todos los graficos relacionales

```python
sns.relplot(data=tips,x='total_bill',y='tip')
plt.show()
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',kind='line')
plt.show()
```

```python
plt.figure(figsize=(12,7))
sns.relplot(data=tips,x='total_bill',y='tip',hue='day',alpha=0.5,col='time')
#Movemos la leyenda
plt.legend(loc='center',bbox_to_anchor=(1.10,0.5))
plt.show()
```
